```python
def analyze_stock(stock_code):
    analyzer = StockAnalyzer()
    report = analyzer.analyze_stock(stock_code)
    prompt = f"""
我打算以最新价格买入，在三个交易日内卖出。分析以下股票是否能够买入，返回结果可以是1:立即买入、-1:风险高于收益，不适合买入、0:当前不是买点需要观望这三种，请不要返回其他内容,仅返回对应的数字。    
    {report}
    """
    client = ZhipuAiClient(api_key="554b0440621a402ba40c9ee3de113f29.wGxg4Y9PHYJ0D8BE")  # 请填写您自己的 API Key
    response = client.chat.completions.create(
        model ="glm-4.5-flash",
        messages=[
            {"role": "system", "content": "你是一个专业的中国股票分析师。"},
            {"role": "user", "content": prompt}
        ],
        thinking={
            "type": "disabled",    # 启用深度思考模式
        },
        max_tokens=4096,          # 最大输出 tokens
        temperature=0.6           # 控制输出的随机性
    )
    return response.choices[0].message.content

```

```python
from get_sock_list import get_hs300_main_board_stocks
stock_list = get_hs300_main_board_stocks()
```

```python
# 获取换手率大于2的股票代码
codes = stock_list[stock_list['换手率(%)'] > 3]['股票代码'].tolist()
print(codes)
```

```python
# 股票代码Python列表（字符串类型，保留前导0）
stock_codes = codes

import time

results = {}

for code in stock_codes:
    print(f"正在分析股票：{code}")
    try:
        result = analyze_stock(code)
        print(f"{code} 分析结果：{result}")
        results[code] = result
    except Exception as e:
        print(f"{code} 分析出错：{e}")
        results[code] = f"分析出错：{e}"
    # 每2分钟分析一次
    time.sleep(120)

print("所有股票分析完成。")
print("结果汇总：")
for code, res in results.items():
    print(f"{code}: {res}")
```
